```python
X = np.linspace(0,4)

@np.vectorize
def func(x):
    return x*x


def myintegrate(func, x0, x1, h=1):
    steps = int(np.abs(x1-x0) / h)
    norm = 1/h
    X = np.linspace(x0, x1, steps)
    integral = func(X).sum() / norm
    return integral

myintegrate(func, 0, 4, h=0.00001)
```
