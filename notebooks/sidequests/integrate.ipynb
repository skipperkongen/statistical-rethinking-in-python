```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt
```

## Test case

```python
# Range we will integrate over
X = np.linspace(-2,4,100)

@np.vectorize
def func(x):
    """
    Quadratic function
    """
    return x*x
```

```python
width = 0.1*1/(X.max()-X.min())
plt.bar(X, func(X), color='red', width=width)
plt.plot(X, func(X))
plt.title('Integral')
```

## Analytical answer

```python
analytical = 4**3 / 3.
print(analytical)
```

## Naive 1

Describe...

```python
X_step = np.linspace(X.min(), X.max(), int(X.max()-X.min()))
naive_1 = func(X_step).sum()
print(naive_1)
```

## Naive 2

Describe...

```python
def naive_integrator(func, x0, x1, h=1):
    steps = int(np.abs(x1-x0) / h)
    norm = 1/h
    X = np.linspace(x0, x1, steps)
    integral = func(X).sum() / norm
    return integral

naive_2 = naive_integrator(func, 0, 4, h=0.00001)
print(naive_2)
```

## SciPy numerical

```python
scipy_numerical = integrate.quad(func, 0, 4)[0]
print(scipy_numerical)
```

## Errors

Show errors between analytical and numerical

```python
plt.bar(
    [
        'Naive1', 
        'Naive2', 
        'Scipy'], 
    [
        naive_1 - analytical,
        naive_2 - analytical,
        scipy_numerical - analytical
    ])
```
