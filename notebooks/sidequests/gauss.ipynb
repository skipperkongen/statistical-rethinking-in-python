## Understanding the normal distribution

Inspired by [3blue1brown: Why Ï€ is in the normal distribution (beyond integral tricks)](https://youtu.be/cy8r7WSuT1I).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
@np.vectorize
def bell1(x):
    return np.exp(-x**2)

@np.vectorize
def bell2(x):
    return np.exp(-x**2) / np.sqrt(np.pi)

@np.vectorize
def gauss(x, sigma=1, mean=0):
    return np.exp(-0.5 * ((x-mean)/sigma)**2) / (sigma * np.sqrt(2* np.pi))

X = np.linspace(-5, 5, 100)
plt.plot(X, bell1(X), label='Bell 1') 
plt.plot(X, bell2(X), label='Bell 2')
plt.plot(X, gauss(X, mean=2, sigma=1.5), label='Gauss') 
plt.legend()
plt.show()
```
